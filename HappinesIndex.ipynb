```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('report_2018-2019.csv', sep=',')
dataset
```

```python
dataset.info()
```

```python
# Qual o país mais feliz?

dataset[dataset['Overall rank'] == 1]
```

```python
# Qual o país menos feliz?

dataset[dataset['Overall rank'] == 156]
```

```python
plt.figure(figsize=(15,8))

# Plot the orbital period with horizontal boxes
sns.boxplot(x="Score", data=dataset,
            whis=[0, 100], width=.6, palette="vlag"); 

# Add in points to show each observation
sns.stripplot(x="Score", data=dataset,
              size=4, color="0", linewidth=0);

plt.show()
```
