# Regression discontinuity design

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Avaliando Polinomial com Score de Felicidade
 
mod = smf.ols(formula = "Score ~ Healthy_life_expectancy + GDP_per_capita + Social_support + I(Score > 3.50)", data=dataset)
res = mod.fit()
print(res.summary())
```

```python
# Plotting fit

# Getting standard errors
prob_se = res.get_prediction().summary_frame()
prob_se['Score'] = dataset['Score']
prob_se.sort_values(by='Score',inplace=True,ignore_index=True)

low = prob_se[prob_se['Score'] <= 3.50].copy()
high = prob_se[prob_se['Score'] > 3.50].copy()
```

```python
prob_se
```

```python
plt.plot(low);
```

```python
plt.plot(high);
```

```python
# Getting effect for threshold bump
coef = res.summary2().tables[1]
ci = coef.iloc[1,4:6].astype(int).to_list()

fig, ax = plt.subplots()

ax.scatter(dataset['Score'], dataset['Score'], c='grey',
           edgecolor='k', alpha=0.15, s=5, zorder=1)

ax.axvline(3.50, linestyle='solid', alpha=1.0, 
           color='k',linewidth=1, zorder=2)

ax.fill_between(low['Score'],low['mean_ci_lower'],
                low['mean_ci_upper'],alpha=0.6,
                zorder=3, color='darkblue')

ax.fill_between(high['Score'],high['mean_ci_lower'],
                high['mean_ci_upper'],alpha=0.6,
                zorder=3, color='red')

ax.set_xlabel('Independent variables')
ax.set_ylabel('Happiness Score')
ax.set_title(f'RDD Reduced Cost Estimate {ci[0]} to {ci[1]} (95% CI)')

# ax.text(0.20, 0.20, 'Threshold', rotation=90, size=9,
#          ha="center", va="center",
#          bbox=dict(boxstyle="round", ec='k',fc='grey'))
```
